```python
# Implementation of the Backpropagation using keras
# Regression based problem
# Author: Muhammad Humayun Khan

df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]],columns = ['cgpa','profile_score','salary'])
df
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,InputLayer
```

```python
model = Sequential()

model.add(InputLayer(shape=(2,)))
model.add(Dense(2, activation='linear'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
# randomly initialize weights
model.get_weights()
```

```python
new_weights = [np.array([[0.1,0.1],[0.1,0.1]],dtype = np.float32),
              np.array([0.,0.], dtype=np.float32),
              np.array([[0.1],
                       [0.1]],dtype=np.float32),
              np.array([0.],dtype=np.float32)]
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimizer = keras.optimizers.Adam(learning_rate = 0.001)
model.compile(loss = 'mean_squared_error',optimizer = optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['salary'].values,epochs = 75,verbose = 1, batch_size = 1)
```
